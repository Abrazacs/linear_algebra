---

**Решение**


Представим векторы в виде матрицы


\begin{pmatrix}
1 & -2 & 4 \\
8 & 3 & -11 \\
-1 & 3 & 9
\end{pmatrix}

Для проверки линейной независимости вычислим ранг этой матрицы. Если ранг матрицы равен числу векторов (т.е. 3), то векторы линейно независимы. Если ранг меньше 3, то векторы линейно зависимы.

Приведем ее к ступенчатому виду.

Вычтем первую строку из второй (предварительно умножив на 8), а также
сложим первую и третью сторки

\begin{pmatrix}
1 & -2 & 4 \\
0 & 19 & -43 \\
0 & 1 & 13
\end{pmatrix}

Вторую сторку делим на 19 и вычетаем из третьей

\begin{pmatrix}
1 & -2 & 4 \\
0 & 19 & -43 \\
0 & 0 & 15
\end{pmatrix}

*15 - это округленное значение*

Ранг ступенчатой матрицы равен количеству её ненулевых строк

Т.к. все строки не нулевые, то ранг = 3.

Следовательно векторы линейно независимы.

**Задание 10**

Разделите данные на предикторы и целевую переменную. Обучите модель с помощью классической линейной регрессии. Линейную регрессию необходимо реализовать без готовых библиотек.

Оцените качество модели с помощью хотя бы двух метрик качества.

Распишите решение с помощью кода на языке Python и напишите текстом интерпретацию полученного вывода.

```python
X = df.drop(['cnt'], axis=1)
y = df['cnt']

X = np.c_[np.ones(X.shape[0]), X]

X = np.array(X)
y = np.array(y)

coefficients = np.linalg.inv(X.T @ X) @ X.T @ y

y_pred = X @ coefficients

print(f'MAE: {metrics.mean_absolute_error(y, y_pred)}')
print(f'MSE: {metrics.mean_squared_error(y, y_pred)}')
print(f'R2: {metrics.r2_score(y, y_pred)}')

```
