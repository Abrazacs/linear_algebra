```python
import numpy as np
import pandas as pd

df = pd.read_csv('data_hw3.csv', sep=',', encoding='cp1251')
df.head()
```

# Задание 1

```python
def define_multiplication_order(A,B,C):
  rows_A = len(A)
  columns_A = len(A[0])
  rows_B = len(B)
  columns_B = len(B[0])
  rows_C = len(C)
  columns_C = len(C[0])
  operations_qty_1 = rows_A * columns_A + columns_B * rows_A * columns_B * columns_C
  operations_qty_2 = rows_B * columns_B * columns_C + rows_A * columns_A * columns_C
  if operations_qty_1 <= operations_qty_2:
    return '(AxB)xC'
  else:
    return 'Ax(BxC)'

A = [[1,2]]
B = [[2],[1]]
C = [[5]]

print(define_multiplication_order(A,B,C))
```

# Задание 2


```python
def best_worst(X, Y):
  corr_matrix = np.corrcoef(X, Y, rowvar=False)
  y_corr_line_without_y = corr_matrix[-1][:-1]
  max_idx = np.argmax(np.abs(y_corr_line_without_y))
  min_idx = np.argmin(np.abs(y_corr_line_without_y))
  return max_idx, min_idx

X = df.loc[:, ['Комнаты',	'Площадь',	'Этаж',	'Центр?',	'Спальный1?',	'Спальный 2?']]
Y = df.loc[:, ['Цена']]
best_worst(X, Y)
```

#Задание 3


```python
def corr_rank(X):
  rank = np.linalg.matrix_rank(
      np.corrcoef(X.T, rowvar=False)
  )
  return rank

print(corr_rank(X))
```

# Задание 4

```python
def inverse_matrix(A):
  det = np.linalg.det(A)
  if abs(det) < (0.1 * 10 ** (-7)):
    return None
  else:
    return np.linalg.inv(A)

A = np.array([[1, 2], [2, 1]])

print(inverse_matrix(A))
```

# Задание 5

```python
def fit_model(X, y):
  return inverse_matrix(X.T @ X) @ X.T @ y

fit_model(X, Y)
```
